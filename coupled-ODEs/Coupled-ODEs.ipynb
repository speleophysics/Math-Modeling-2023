```python
# imports
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# define parameters and initial conditions
k = 1
m = 0.5
x0 = 1.
V0 = 0.

# Create an array to hold time values
t = np.linspace(0, 10, 500)

# Define derivative function
def dy_dt(y, t):
    x = y[0]
    V = y[1]
    dx_dt = V
    dV_dt = -(k/m)*x
    return [dx_dt, dV_dt]

# Solve with odeint
y = odeint(dy_dt, [x0, V0], t)

# Plot solution
x_sol = y[:,0]
V_sol = y[:,1]
plt.plot(t, x_sol)
plt.plot(t, V_sol, '--')
plt.legend(['x', 'V'])
```
